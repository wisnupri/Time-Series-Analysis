```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as py
import plotly.offline
import plotly.graph_objects as go

%matplotlib inline
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/UTS 2022 Semester1/Curah Jawa Tengah.csv")
df.head()
```

```python
df.describe()
```

```python
# prepare dataframes for plots
df_month = df[['TAHUN', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']]
```

## **Curah Hujan Tahunan yang diterima di wilayah Jawa Tengah**

```python
# Memvisualisasikan curah hujan tahunan selama (2017-2021) di Jawa Tengah
df.groupby("TAHUN").mean()['ANNUAL'].plot(ylim=(800,3000),color='k',marker='o',markerfacecolor='red',linestyle='-',linewidth=2,figsize=(12,10))
plt.xlabel('Tahun',fontsize=20)
plt.ylabel('Curah Hujan Tahunan (dalam mm)',fontsize=20)
plt.title('Curah Hujan Tahunan dari Tahun 2017 hingga 2021 di Jawa Tengah',fontsize=25)

plt.grid()
```

*Dapat di lihat dari Grafik diatas Perolehan curah hujan di wilayah Jawa Tengah dari Tahun 2017 sampai 2021, perolehan curah hujan terrendah berada di tahun 2018.*

## **Bulan dengan curah hujan tertinggi**

```python
plt.figure(figsize=(10,7))
df[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG',
       'SEP', 'OCT', 'NOV', 'DEC']].mean().plot(kind= 'bar', color='red')
plt.xlabel('Bulan',fontsize=15)
plt.ylabel('Rata-rata. Curah hujan',fontsize=15)
plt.title('Rata-rata Data Curah Hujan Bulanan',fontsize=25)
plt.grid()
plt.show()
```

```python
Z = df[['TAHUN', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL','AUG', 'SEP', 'OCT', 'NOV', 'DEC']].sort_values('TAHUN').groupby("TAHUN").mean().reset_index()
ax = Z[['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL','AUG', 'SEP', 'OCT', 'NOV', 'DEC']].plot(figsize=(13,8))
YL = df.sort_values('TAHUN')["TAHUN"].unique().tolist()
plt.xticks(range(0,5), YL, fontsize=8, rotation=90)
plt.show()
```

## **Heat Map Curah Hujan**

```python
fig=plt.gcf()
fig.set_size_inches(15,15)
fig=sns.heatmap(df.corr(),annot=True,linewidths=1,linecolor='k',square=True,mask=False, vmin=-1, vmax=1,cbar_kws={"orientation": "vertical"},cbar=True)
```

Heatmap merupakan teknik untuk mengetahui perilaku curah hujan di datasets dengan menyediakan data berupa statistik. Dengan tools ini kita bisa melihat area mana dari data memiliki nilai tertinggi yang ditunjukkan dengan warna gelap sehingga kita bisa tahu di bulan apa.

Heatmap ini menunjukkan koefisien korelasi curah hujan bulanan antar bulan. Jelas tidak ada korelasi yang signifikan antara bulan. Contoh Hanya korelasi 0,68 antara Juni dan April terlihat lebih tinggi, tetapi masih terlalu rendah. Secara umum, peta panas mengatakan memprediksi curah hujan bulanan dengan korelasi tidak cukup ideal.
